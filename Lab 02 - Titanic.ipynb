```python
conditions = [
    (df['Fare'] <= 7.92500),
    (df['Fare'] > 7.92500) & (df['Fare'] <= 14.45420),
    (df['Fare'] > 14.45420) & (df['Fare'] <= 31.13750),
    (df['Fare'] > 31.13750)]
choices = ['1', '2', '3', '4']
df['FareGroup'] = np.select(conditions, choices)
```

```python
sns.countplot(x = 'FareGroup', hue='Survived', data=df)
```

```python
df.groupby('FareGroup').size()
```

### Difference in survival rates based on Fare using a graph

- To determine the survival rate based on Fare, I classified the people based on the Fare percentile
> - Group 1 - Less or equal to 25% of the Fare
  - Group 2 - More than 25% and less than or equal to 50% of the Fare
  - Group 3 - More than 50% and less than or equal to 75% of the Fare
  - Group 4 - More than 75% of the Fare
- From the graph you can see that group 1, 2, and 3 has less people who survive, because they pay less money. While group 4 has more people who survived than not survived.

```python
sns.catplot(x="FareGroup", hue="Sex", col="Survived",
                data=df, kind="count",
                height=4, aspect=.7);
```

Using the Sex column, we can even see the information that no matter which Fare Group, female had survived more than male.
