### Data Inspection

- There is no null field in the dataset
- To show the relationship between the fare paid and survival, Survived and Fare columns are required
- Survived - there are more people who died than who suvived as the mean is 0.3856
- Fare - there are more people who paid lower fare
- Age - this column could be used if we want to know whether younger or older people has survived
- Pclass - this can be used for grouping of the people who pays lower or higher fare

```python
df.groupby('Pclass').mean()
```

```python
df[['Survived', 'Fare']].head(10)
```

```python
p1_df = df.loc[df['Pclass'] == 1]
p1_df.head()
```

```python
p1_survived_rate = p1_df['Survived'].value_counts(normalize=True)
p1_survived_rate
```

```python
p2_df = df.loc[df['Pclass'] == 2]
p2_df.head()
```

```python
p2_survived_rate = p2_df['Survived'].value_counts(normalize=True)
p2_survived_rate
```

```python
p3_df = df.loc[df['Pclass'] == 3]
p3_df.head()
```

```python
p3_survived_rate = p3_df['Survived'].value_counts(normalize=True)
p3_survived_rate
```

```python
sns.countplot(x = 'Pclass', hue='Survived', data=df)
```

- While Pclass 2, second class, has more peole who didn't survive and Pclass 3, third class, has the most peopele who didn't survived.
